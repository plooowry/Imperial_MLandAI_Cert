## üîπ Representation of a Vector

A **vector** can be represented using boldface letters (e.g., **v**) or as a list of numerical components:
$$
\mathbf{v} = (v_1, v_2, v_3)
$$

Vectors are widely used in machine learning to represent:
- **Input features** of data points
- **Output labels** or **predictions**

Each component of a vector corresponds to a distinct feature or attribute.

Example 2D vector representation:
- $\mathbf{v} = (v_1, v_2)$ is plotted with $v_1$ and $v_2$ as components along the x and y axes respectively.

---

## üîπ Basic Vector Operations

- **Vector addition:**  
  Add two vectors component-wise:
  $$
  \mathbf{a} = (a_1, a_2, a_3), \quad \mathbf{b} = (b_1, b_2, b_3)
  $$
  $$
  \mathbf{a} + \mathbf{b} = (a_1 + b_1, a_2 + b_2, a_3 + b_3)
  $$

- **Scalar multiplication:**  
  Multiply each component of a vector by a scalar $k$:
  $$
  \mathbf{v} = (v_1, v_2, v_3) \Rightarrow k \cdot \mathbf{v} = (kv_1, kv_2, kv_3)
  $$

- **Vector multiplication:**  
  Two common types:
  - **Dot product** (returns scalar):  
    $$
    \mathbf{a} \cdot \mathbf{b} = a_1b_1 + a_2b_2 + a_3b_3
    $$
  - **Cross product** (returns vector, for 3D):  
    $$
    \mathbf{a} \times \mathbf{b} = 
    \begin{vmatrix}
    \mathbf{i} & \mathbf{j} & \mathbf{k} \\
    a_1 & a_2 & a_3 \\
    b_1 & b_2 & b_3 \\
    \end{vmatrix}
    = (a_2b_3 - a_3b_2)\mathbf{i} - (a_1b_3 - a_3b_1)\mathbf{j} + (a_1b_2 - a_2b_1)\mathbf{k}
    $$

- **Vector subtraction:**  
  Subtract component-wise:
  $$
  \mathbf{a} - \mathbf{b} = (a_1 - b_1, a_2 - b_2, a_3 - b_3)
  $$

---

## üîπ Operation Example

Let:
- $\mathbf{A} = (3, 4, 5)$  
- $\mathbf{B} = (2, 1, 6)$

### ‚ûï Vector Addition
$$
\mathbf{A} + \mathbf{B} = (3 + 2, 4 + 1, 5 + 6) = (5, 5, 11)
$$

### ‚ûñ Vector Subtraction
$$
\mathbf{A} - \mathbf{B} = (3 - 2, 4 - 1, 5 - 6) = (1, 3, -1)
$$

### ‚úñ Scalar Multiplication
Let $k = 2$:
$$
2\mathbf{A} = (2 \cdot 3, 2 \cdot 4, 2 \cdot 5) = (6, 8, 10)
$$

---

## üî∏ Dot Product (Scalar Product)

Let:
- $\mathbf{A} = (a_1, a_2, a_3) = (3, 4, 5)$  
- $\mathbf{B} = (b_1, b_2, b_3) = (2, 1, 6)$

Then:
$$
\mathbf{A} \cdot \mathbf{B} = a_1b_1 + a_2b_2 + a_3b_3 \\
= (3 \cdot 2) + (4 \cdot 1) + (5 \cdot 6) = 6 + 4 + 30 = 40
$$

---

## üî∏ Cross Product

Same vectors:
$$
\mathbf{A} = (3, 4, 5), \quad \mathbf{B} = (2, 1, 6)
$$

Cross product:
$$
\mathbf{A} \times \mathbf{B} = 
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
3 & 4 & 5 \\
2 & 1 & 6 \\
\end{vmatrix}
= (4 \cdot 6 - 5 \cdot 1)\mathbf{i} - (3 \cdot 6 - 5 \cdot 2)\mathbf{j} + (3 \cdot 1 - 4 \cdot 2)\mathbf{k}
= (24 - 5)\mathbf{i} - (18 - 10)\mathbf{j} + (3 - 8)\mathbf{k}
= 19\mathbf{i} - 8\mathbf{j} - 5\mathbf{k}
$$

---

‚úÖ These operations are foundational for expressing computations in ML models like gradient descent, transformations, feature projections, and neural networks.

## Matrix Addition and Subtraction

Matrices must have the same dimensions to add or subtract them. The operation is performed by adding or subtracting the corresponding elements.

Consider two matrices $A$ and $B$:
$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} \quad B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
$$

The sum is:
$$
A + B = \begin{bmatrix} 1+5 & 2+6 \\ 3+7 & 4+8 \end{bmatrix} = \begin{bmatrix} 6 & 8 \\ 10 & 12 \end{bmatrix}
$$

The difference is:
$$
A - B = \begin{bmatrix} 1-5 & 2-6 \\ 3-7 & 4-8 \end{bmatrix} = \begin{bmatrix} -4 & -4 \\ -4 & -4 \end{bmatrix}
$$

## Matrix-Scalar Multiplication

This involves multiplying every element of a matrix by a single numerical value (a scalar).

If $ k = 3 $ and $ A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} $, then:
$$
kA = 3 \times \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} = \begin{bmatrix} 3 \times 1 & 3 \times 2 \\ 3 \times 3 & 3 \times 4 \end{bmatrix} = \begin{bmatrix} 3 & 6 \\ 9 & 12 \end{bmatrix}
$$

## Matrix Multiplication (Dot Product)

The number of **columns** in the first matrix must equal the number of **rows** in the second. Each element in the resulting matrix is the dot product of a row from the first matrix and a column from the second.

Consider matrices $A$ (a $2 \times 3$ matrix) and $B$ (a $3 \times 2$ matrix):
$$
A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix} \quad B = \begin{bmatrix} 7 & 8 \\ 9 & 10 \\ 11 & 12 \end{bmatrix}
$$

The resulting matrix $C = A \times B$ will have dimensions $2 \times 2$.

* **Element C11 (Row 1, Col 1):** (1 &middot; 7) + (2 &middot; 9) + (3 &middot; 11) = 7 + 18 + 33 = 58
* **Element C12 (Row 1, Col 2):** (1 &middot; 8) + (2 &middot; 10) + (3 &middot; 12) = 8 + 20 + 36 = 64
* **Element C21 (Row 2, Col 1):** (4 &middot; 7) + (5 &middot; 9) + (6 &middot; 11) = 28 + 45 + 66 = 139
* **Element C22 (Row 2, Col 2):** (4 &middot; 8) + (5 &middot; 10) + (6 &middot; 12) = 32 + 50 + 72 = 154

Resulting matrix:
$$
C = \begin{bmatrix} 58 & 64 \\ 139 & 154 \end{bmatrix}
$$

## Special Types of Matrices and Their Properties

Understanding these special matrices and properties is crucial for linear algebra and its applications in machine learning.

---

### Special Matrices

**1. Identity Matrix ($I$)**

An identity matrix is a square matrix with **ones on the main diagonal** and **zeros everywhere else**. It acts as the multiplicative identity, meaning that for any matrix $A$, the product $A \cdot I = A$ and $I \cdot A = A$.

$$I = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}$$

**2. Zero Matrix**

A zero matrix (or null matrix) is a matrix where **all elements are zero**. It's the additive identity, meaning $A + 0 = A$.

$$\begin{bmatrix} 0 & 0 & 0 \\ 0 & 0 & 0 \\ 0 & 0 & 0 \end{bmatrix}$$

---

### Key Matrix Properties

These properties govern how matrix operations work:

* **Associative Property (Multiplication):** The way matrices are grouped in multiplication doesn't change the result.
    $$(A \cdot B) \cdot C = A \cdot (B \cdot C)$$

* **Distributive Property:** Matrix multiplication distributes over matrix addition.
    $$A \cdot (B + C) = A \cdot B + A \cdot C$$

* **Non-commutative Property (Multiplication):** The order of matrix multiplication matters! In general, multiplying $A$ by $B$ is not the same as multiplying $B$ by $A$. This is a major difference from scalar multiplication. ‚öñÔ∏è
    $$A \cdot B \neq B \cdot A$$
   

## Mini-Lesson: Eigenvalues and Eigenvectors üß†

Imagine you're in a field of flagpoles on a windy day. The wind acts as a transformation. Most flagpoles lean and sway, changing their direction. However, a few special flagpoles might only stretch or flip along their original direction.

* **Eigenvectors** are these special flagpoles‚Äîthey are vectors that don't change direction when a linear transformation is applied, they only scale.
* **Eigenvalues** are the numbers that tell you *how much* an eigenvector stretches or flips.

The mathematical representation of this relationship is:

$$A\vec{v} = \lambda\vec{v}$$

Where:
* $A$ is the matrix representing the transformation.
* $\vec{v}$ is the eigenvector (the direction that doesn't change).
* $\lambda$ is the scalar eigenvalue (the amount of stretch or flip).

---

### How to Compute Eigenvalues and Eigenvectors

To find the eigenvalues and eigenvectors for a square matrix $A$, we follow these steps.

**Step 1: Form the Characteristic Equation**

The core idea is to find a non-zero vector $\vec{v}$ that satisfies $(A - \lambda I)\vec{v} = 0$. This equation has a non-trivial solution only if the matrix $(A - \lambda I)$ is singular, which means its determinant is zero.

This gives us the **characteristic equation**:

$$\det(A - \lambda I) = 0$$

> **What is a determinant?** A determinant is a scalar value computed from a square matrix. For a 2x2 matrix, the formula is:
> $$ \det \begin{pmatrix} a & b \\ c & d \end{pmatrix} = ad - bc $$
> If the determinant is zero, the matrix is singular (non-invertible).

**Step 2: Apply the Characteristic Equation to an Example**

Let's use the matrix $A = \begin{bmatrix} 2 & 1 \\ 1 & 1 \end{bmatrix}$.

First, set up the expression for $\det(A - \lambda I)$:
$$ \det \left( \begin{bmatrix} 2 & 1 \\ 1 & 1 \end{bmatrix} - \lambda \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} \right) = \det \begin{pmatrix} 2-\lambda & 1 \\ 1 & 1-\lambda \end{pmatrix} = 0 $$

Now, calculate the determinant:
$$ (2-\lambda)(1-\lambda) - (1)(1) = 0 $$

**Step 3: Solve for the Eigenvalues ($\lambda$)**

Simplify the polynomial equation from the previous step:
$$ 2 - 2\lambda - \lambda + \lambda^2 - 1 = 0 $$
$$ \lambda^2 - 3\lambda + 1 = 0 $$

We can solve for $\lambda$ using the quadratic formula, $\lambda = \frac{-b \pm \sqrt{b^2-4ac}}{2a}$:
$$ \lambda = \frac{-(-3) \pm \sqrt{(-3)^2 - 4(1)(1)}}{2(1)} = \frac{3 \pm \sqrt{9-4}}{2} = \frac{3 \pm \sqrt{5}}{2} $$

The two eigenvalues are:
$$ \lambda_1 = \frac{3+\sqrt{5}}{2} \approx 2.618 \quad \text{and} \quad \lambda_2 = \frac{3-\sqrt{5}}{2} \approx 0.382 $$

**Step 4: Compute the Eigenvectors ($\vec{v}$)**

For each eigenvalue $\lambda$, solve the system of linear equations $(A - \lambda I)\vec{v} = 0$ to find the corresponding eigenvector $\vec{v}$.

For this example, the computed eigenvectors are:
$$ \vec{v_1} \approx \begin{bmatrix} 0.851 \\ 0.526 \end{bmatrix} \quad (\text{for } \lambda_1) \qquad \vec{v_2} \approx \begin{bmatrix} -0.526 \\ 0.851 \end{bmatrix} \quad (\text{for } \lambda_2) $$

## Mini-Lesson: Linear Transformations üìê

A **linear transformation** is a consistent way to map one set of vectors to another. Think of it like a recipe that predictably changes ingredients (vectors) into something new. Common examples include scaling, rotating, or reflecting an image.

---

### Scaling

Scaling involves multiplying an entire vector by a scalar value. This **stretches** or **compresses** the vector without changing its direction.

For example, to scale the vector $\vec{v} = \begin{pmatrix} 3 \\ 2 \end{pmatrix}$ by a scalar $K = 2$:
$$ K \cdot \vec{v} = 2 \begin{pmatrix} 3 \\ 2 \end{pmatrix} = \begin{pmatrix} 6 \\ 4 \end{pmatrix} $$

---

### Reflection

Reflection involves **flipping a vector** across an axis, like looking in a mirror. This is done by multiplying the vector by a reflection matrix.

For example, to reflect the vector $\vec{v} = \begin{pmatrix} 3 \\ 4 \end{pmatrix}$ across the x-axis, we use the transformation matrix $T = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$:
$$ T \vec{v} = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix} \begin{pmatrix} 3 \\ 4 \end{pmatrix} = \begin{pmatrix} 3 \\ -4 \end{pmatrix} $$

---

### Rotation

Rotation **changes a vector's direction** while preserving its length (magnitude). The 2D rotation matrix for an angle $\theta$ is:
$$ R(\theta) = \begin{bmatrix} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{bmatrix} $$

To rotate the vector $\vec{v} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$ by $90^\circ$, we use $R(90^\circ)$, noting that $\cos(90^\circ)=0$ and $\sin(90^\circ)=1$:
$$ R(90^\circ)\vec{v} = \begin{bmatrix} 0 & -1 \\ 1 & 0 \end{bmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \begin{pmatrix} 0 \\ 1 \end{pmatrix} $$

---

### Projection

Projection involves "flattening" a vector onto a line, essentially casting its shadow. This is achieved by multiplying the vector by a projection matrix.

For example, to project the vector $\vec{v} = \begin{pmatrix} 3 \\ 4 \end{pmatrix}$ onto the x-axis, we use the transformation matrix $T = \begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}$:
$$ T \vec{v} = \begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix} \begin{pmatrix} 3 \\ 4 \end{pmatrix} = \begin{pmatrix} 3 \\ 0 \end{pmatrix} $$

## Mini-Lesson: Matrix Decomposition üß©

**Matrix decomposition** is a powerful technique that breaks down complex matrices into simpler, constituent parts. This makes them easier to analyze and is highly efficient for solving problems in machine learning and linear algebra.

---

### Singular Value Decomposition (SVD)

SVD is a fundamental technique that decomposes any matrix $A$ into three simpler matrices:

$$ A = U \Sigma V^T $$

Where:
* **$U$** is the matrix of **left singular vectors**, which capture patterns in the rows of $A$.
* **$\Sigma$** (Sigma) is a diagonal matrix containing the **singular values**.
* **$V^T$** is the transpose of the matrix of **right singular vectors**.

#### SVD Example Walkthrough
Let's decompose the matrix $A = \begin{bmatrix} 3 & 0 \\ 4 & 0 \end{bmatrix}$.

**1. Find Eigenvectors and Eigenvalues for $V$ and $U$**
We compute $A^T A$ to find $V$, and $AA^T$ to find $U$.

* $A^T A = \begin{bmatrix} 3 & 4 \\ 0 & 0 \end{bmatrix} \begin{bmatrix} 3 & 0 \\ 4 & 0 \end{bmatrix} = \begin{bmatrix} 25 & 0 \\ 0 & 0 \end{bmatrix}$
* $AA^T = \begin{bmatrix} 3 & 0 \\ 4 & 0 \end{bmatrix} \begin{bmatrix} 3 & 4 \\ 0 & 0 \end{bmatrix} = \begin{bmatrix} 9 & 12 \\ 12 & 16 \end{bmatrix}$

Solving for the eigenvectors of these matrices gives us:
* For $A^T A$, the eigenvectors form $V = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}$.
* For $AA^T$, the eigenvectors form $U = \begin{bmatrix} 3/5 & -4/5 \\ 4/5 & 3/5 \end{bmatrix}$.

**2. Compute Singular Values ($\Sigma$)**
The singular values are the square roots of the non-zero eigenvalues. The only non-zero eigenvalue is 25.
$$ \sigma_1 = \sqrt{25} = 5 $$
We place this in the diagonal matrix $\Sigma$:
$$ \Sigma = \begin{bmatrix} 5 & 0 \\ 0 & 0 \end{bmatrix} $$

**3. Assemble the Final Decomposition**
Finally, we combine $U$, $\Sigma$, and $V^T$. First, we find the transpose of $V$. (In this special case, $V$ is the identity matrix, so its transpose is itself).
$$ V = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} \quad \implies \quad V^T = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} $$
Now we put it all together in the formula $A = U \Sigma V^T$:
$$ A = \begin{bmatrix} 3/5 & -4/5 \\ 4/5 & 3/5 \end{bmatrix} \begin{bmatrix} 5 & 0 \\ 0 & 0 \end{bmatrix} \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} $$

---

### Lower-Upper (LU) Decomposition

LU decomposition expresses a square matrix $A$ as the product of a **lower triangular matrix ($L$)** and an **upper triangular matrix ($U$)**.

$$ A = L \cdot U $$

#### LU Example
Let's decompose the matrix $A = \begin{bmatrix} 4 & 3 \\ 6 & 3 \end{bmatrix}$. The resulting $L$ and $U$ matrices are:
$$ L = \begin{bmatrix} 1 & 0 \\ 1.5 & 1 \end{bmatrix} \quad \text{and} \quad U = \begin{bmatrix} 4 & 3 \\ 0 & -1.5 \end{bmatrix} $$

Computing Eigenvalues and Eigenvectors

```python
import numpy as np

# Define the matrix A
A = np.array([[1, 2], [3, 4]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues: ", eigenvalues)
print("Eigenvectors: \n", eigenvectors)

# Verify the relationship Av = Œªv for the first eigenvector
lambda1 = eigenvalues[0]
v1 = eigenvectors[:, 0]

# Normalize the eigenvector for easier comparison
v1_normalized = v1 / np.linalg.norm(v1)

# Compute Av
Av = np.dot(A, v1_normalized)

# Compute Œªv
lambda_v = lambda1 * v1_normalized

print("\nVerification:")
print("Av: ", Av)
print("Œªv: ", lambda_v)
```

### Question 1

Given two vectors:
$$ \vec{a} = (2, 5, 6) $$
$$ \vec{b} = (3, 7, 4) $$
Find the cross product $\vec{a} \times \vec{b}$.

* A) $(22, -10, -1)$
* B) $(-22, 10, -1)$
* C) $(-22, 10, 1)$
* D) $(-10, -22, 1)$

---

### Solution with Working ‚úçÔ∏è

The cross product $\vec{a} \times \vec{b}$ can be calculated using the determinant of a matrix formed by the unit vectors $\mathbf{i}, \mathbf{j}, \mathbf{k}$ and the components of $\vec{a}$ and $\vec{b}$.

**1. Set up the determinant:**

$$
\vec{a} \times \vec{b} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
2 & 5 & 6 \\
3 & 7 & 4
\end{vmatrix}
$$

**2. Calculate the components:**

* **i component (x-axis):**
    $$ (5 \cdot 4 - 6 \cdot 7) = 20 - 42 = -22 $$

* **j component (y-axis):**
    $$ (6 \cdot 3 - 2 \cdot 4) = 18 - 8 = 10 $$

* **k component (z-axis):**
    $$ (2 \cdot 7 - 5 \cdot 3) = 14 - 15 = -1 $$

**3. Combine the components:**

The resulting vector is $(-22, 10, -1)$.

**Final Answer:** The correct option is **B) $(-22, 10, -1)$**. ‚úÖ

### ------------------------ Calculus

## Derivatives and Partial Derivatives üìà

This note covers the fundamentals of derivatives for single-variable functions and extends the concept to partial derivatives for multivariable functions.

---

### What is a Derivative?

A derivative measures the **rate of change** of a function with respect to its variables. Geometrically, it represents the **slope of the tangent line** to the function's graph at a specific point.

Its formal mathematical representation is the limit definition:
$$ f'(x) = \lim_{h\to0} \frac{f(x+h) - f(x)}{h} $$

Calculus uses a set of standard rules for differentiating common functions.

#### Common Functions and Their Derivatives

| Name of Function | Function | Derivative | Example |
| :--- | :--- | :--- | :--- |
| **Polynomial** | $f(x) = ax^n$ | $f'(x) = n \cdot ax^{n-1}$ | $f(x) = 5x^2 \implies f'(x) = 10x$ |
| **Exponential** | $f(x) = e^x$ | $f'(x) = e^x$ | $f(x) = e^x \implies f'(x) = e^x$ |
| **Logarithmic** | $f(x) = \ln(x)$ | $f'(x) = \frac{1}{x}$ | $f(x) = \ln(2x) \implies f'(x)=\frac{1}{x}$ |
| **Trigonometric** | $f(x) = \sin(x)$ | $f'(x) = \cos(x)$ | |
| | $f(x) = \cos(x)$ | $f'(x) = -\sin(x)$ | |
| | $f(x) = \tan(x)$ | $f'(x) = \sec^2(x)$ | |
| **Inverse Trig** | $f(x) = \arcsin(x)$ | $f'(x) = \frac{1}{\sqrt{1-x^2}}$ | |
| | $f(x) = \arctan(x)$ | $f'(x) = \frac{1}{1+x^2}$ | |
| **Rational** | $f(x) = \frac{u(x)}{v(x)}$ | $f'(x) = \frac{u'(x)v(x) - u(x)v'(x)}{v(x)^2}$ | |

---

### Partial Derivatives

Partial derivatives extend the concept to functions with multiple variables, like $z = f(x, y)$. They measure the rate of change with respect to **one variable** while holding the **other variables constant**.

Think of it like finding the slope of the function in one specific direction (e.g., parallel to the x-axis or y-axis).

* The partial derivative with respect to $x$ is written as $\frac{\partial z}{\partial x}$. This tells you how $z$ changes if only $x$ changes.
* The partial derivative with respect to $y$ is written as $\frac{\partial z}{\partial y}$. This tells you how $z$ changes if only $y$ changes.

The formal definition for the partial derivative with respect to $x$ is:
$$ f_x(x, y) = \lim_{h\to0} \frac{f(x+h, y) - f(x, y)}{h} $$
This measures how the function changes with respect to one variable while keeping the other(s) constant.

## Mini-Lesson: The Chain Rule ‚õìÔ∏è

The **Chain Rule** is a fundamental method for finding the derivative of a **composite function**‚Äîa function nested inside another, like $h(x) = f(g(x))$. It states that the derivative of a composite function is the product of the derivatives of its component functions.

---

### The Rule

There are two common ways to write the Chain Rule.

1.  **Using Prime Notation:** For a function $h(x) = f(g(x))$, the derivative is:
    $$h'(x) = f'(g(x)) \cdot g'(x)$$
    In words: "the derivative of the outer function (with the inner function left inside) times the derivative of the inner function."

2.  **Using Leibniz's Notation:** If you have a chain of functions, like $z = f(y)$ and $y = g(x)$, the derivative of $z$ with respect to $x$ is:
    $$\frac{dz}{dx} = \frac{dz}{dy} \cdot \frac{dy}{dx}$$

---

### Example

Let's find the derivative of the composite function $z = \sin(x^2)$.

We can break this down into two functions:
* $z = \sin(y)$
* $y = x^2$

**1. Find the derivative of each part:**

* The derivative of $z$ with respect to $y$:
    $$\frac{dz}{dy} = \cos(y)$$
* The derivative of $y$ with respect to $x$:
    $$\frac{dy}{dx} = 2x$$

**2. Apply the Chain Rule:**

Substitute these into the Chain Rule formula:
$$\frac{dz}{dx} = \frac{dz}{dy} \cdot \frac{dy}{dx} = \cos(y) \cdot 2x$$

**3. Substitute back for the original variable:**

Since we want the final answer in terms of $x$, replace $y$ with $x^2$:
$$\frac{dz}{dx} = 2x \cdot \cos(x^2)$$

## Vector Norms üìè

A **norm** is a function that assigns a strictly positive length or "size" to a vector. It's a way to measure the magnitude of a vector. The type of norm is determined by a value 'p'.

---

### $L_1$ Norm (Manhattan Norm)
When $p=1$, the norm is called the **Manhattan** or **$L_1$ norm**. It's calculated as the sum of the absolute values of the vector's components. Think of it as the distance you'd travel in a city grid (like Manhattan), where you can only move along the axes.

$$ \|\vec{x}\|_1 = |x_1| + |x_2| + \dots + |x_n| $$

---

### $L_2$ Norm (Euclidean Norm)
When $p=2$, the norm is the familiar **Euclidean** or **$L_2$ norm**. It represents the shortest, straight-line distance between two points (the length of the vector). It's calculated as the square root of the sum of the squared components.

$$ \|\vec{x}\|_2 = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2} $$

---

### $L_\infty$ Norm (Maximum Norm)
When $p = \infty$, it is called an **infinity norm** or **Maximum norm**. This norm simply takes the largest absolute value among the vector's components.

$$ \|\vec{x}\|_\infty = \max(|x_1|, |x_2|, \dots, |x_n|) $$

---

### Worked Example

Let's calculate all three norms for the vector $\vec{x} = (3, -4)$:

* **$L_1$ norm**:
    $$ \|\vec{x}\|_1 = |3| + |-4| = 3 + 4 = 7 $$
* **$L_2$ norm**:
    $$ \|\vec{x}\|_2 = \sqrt{3^2 + (-4)^2} = \sqrt{9 + 16} = \sqrt{25} = 5 $$
* **$L_\infty$ norm**:
    $$ \|\vec{x}\|_\infty = \max(|3|, |-4|) = \max(3, 4) = 4 $$

---

### Relationship Between Norms

The p-norms are related by the following inequality:
$$ \|\vec{x}\|_p \le \|\vec{x}\|_q \quad \text{whenever} \quad q \ge p $$
In simple terms, as you increase the value of $p$, the calculated norm value gets **smaller** or stays the same.

We can see this in our example:
$$ \|\vec{x}\|_1 \ge \|\vec{x}\|_2 \ge \|\vec{x}\|_\infty $$
$$ 7 \ge 5 \ge 4 $$

Let $f(x, y) = 4xy^2 + x^2$. What is the partial derivative of $f$ with respect to $y$ at the point $(x = 1, y = 2)$?

### Solution with Working üß†

The process involves two steps: first finding the partial derivative expression, then evaluating it at the given point.

To find the partial derivative with respect to $y$, we treat the variable $x$ as a constant.

# Probability and Statistics 

## Mini-Lesson: Probability Distributions üìä

Distributions describe the **likelihood** of different outcomes for a random variable. They show how data is spread out by assigning a probability to each possible outcome.

---

### Types of Distributions

There are two main categories of probability distributions:

* **Discrete Distributions:** These are used for **countable** events where the outcomes are distinct and separate.
    * **Example:** The number of website clicks per day (you can have 10 clicks or 11, but not 10.5).
    * **Common Type:** Binomial distribution.

* **Continuous Distributions:** These are used for **continuous** values that can take any value within a given range.
    * **Example:** The time spent on a website (you can spend 30 seconds, 30.1 seconds, 30.11 seconds, etc.).
    * **Common Type:** Normal distribution.

---

### Key Properties

* **Parameters:** A distribution is defined by its specific parameters, most commonly its **mean** (average value) and **variance** (measure of spread).
* **Total Probability:** The sum of all probabilities for every possible outcome in a distribution must always equal **1**.

## Mini-Lesson: Expectation and Variance üéØ

Expectation and variance are two key properties that describe a probability distribution.

---

### Expectation (Mean, $E[X]$)
The **expectation** is the average value or the central tendency of a random variable. It represents the "center of mass" of the distribution.

* **For discrete variables**, it's the sum of each value multiplied by its probability:
    $$ E[X] = \sum x \cdot P(x) $$

* **For continuous variables**, it's the integral of each value multiplied by its probability density function:
    $$ E[X] = \int x \cdot f(x)dx $$

---

### Variance ($\text{Var}(X)$)
The **variance** measures the spread or dispersion of a distribution. It describes how much the values deviate from the mean. A high variance indicates a wide spread, while a low variance indicates that values are clustered closely around the mean.

The formula is the expected squared deviation from the mean:
$$ \text{Var}(X) = E[(X - E[X])^2] $$

For easier calculation, this is often computed as:
$$ \text{Var}(X) = E[X^2] - (E[X])^2 $$

---

### Worked Example
A website tracks the number of user logins per hour. The observed data provides the following probability distribution:
* **1 login:** 0.2 probability
* **2 logins:** 0.3 probability
* **3 logins:** 0.4 probability
* **4 logins:** 0.1 probability

Let's calculate the expected number of logins and the variance.

#### Expected Value Calculation
$$ E[X] = \sum x \cdot P(x) $$
$$ E[X] = (1 \cdot 0.2) + (2 \cdot 0.3) + (3 \cdot 0.4) + (4 \cdot 0.1) $$
$$ E[X] = 0.2 + 0.6 + 1.2 + 0.4 = 2.4 $$
The expected number of logins per hour is **2.4**.

#### Variance Calculation
We'll use the formula $\text{Var}(X) = E[X^2] - (E[X])^2$. First, we need to calculate $E[X^2]$:
$$ E[X^2] = \sum x^2 \cdot P(x) $$
$$ E[X^2] = (1^2 \cdot 0.2) + (2^2 \cdot 0.3) + (3^2 \cdot 0.4) + (4^2 \cdot 0.1) $$
$$ E[X^2] = (1 \cdot 0.2) + (4 \cdot 0.3) + (9 \cdot 0.4) + (16 \cdot 0.1) $$
$$ E[X^2] = 0.2 + 1.2 + 3.6 + 1.6 = 6.6 $$
Now we can find the variance:
$$ \text{Var}(X) = E[X^2] - (E[X])^2 = 6.6 - (2.4)^2 = 6.6 - 5.76 = 0.84 $$
The variance of the number of logins is **0.84**.

## Mini-Lesson: Entropy üé≤

**Entropy** is a key concept in information theory that measures the **uncertainty** or **randomness** in data. It helps data scientists and machine learning models deal with unpredictability.

---

### Understanding Entropy
Entropy increases as outcomes become more random or disordered. A simple analogy is a deck of cards: a freshly shuffled deck has very high entropy, while a deck neatly arranged by suit and number has very low entropy.

In short, **higher entropy means more unpredictability**.

---

### Entropy in Machine Learning
Entropy is essential for optimizing model performance in several ways:
* It forms the basis of **cross-entropy loss**, a critical metric for evaluating classification tasks.
* It guides feature selection and data splits in decision trees through a concept called **information gain**.
* It can help in **anomaly detection** by flagging unusual data patterns.

**Cross-entropy** is a particularly common loss function that measures the difference between two probability distributions: the true labels and the model's predicted probabilities.

---

### Shannon Entropy ($H(X)$)
In information theory, **Shannon entropy** ($H$) quantifies the average uncertainty for a single random variable. For discrete probabilities, you calculate it using the formula:
$$ H(X) = - \sum_{i=1}^{n} p_i \log_2 p_i $$
where $p_i$ is the probability of outcome $i$. The result is often measured in "bits" and represents the average number of bits required to encode an outcome from the distribution.

#### Example: A Fair Six-Sided Die
Let's calculate the Shannon entropy for a fair six-sided die.

1.  **Probabilities:** For a fair die, there are $n=6$ outcomes, and each has an equal probability of occurring:
    $$ p_i = \frac{1}{6} $$

2.  **Calculation:** We plug this into the entropy formula:
    $$ H(X) = - \sum_{i=1}^{6} \frac{1}{6} \log_2\left(\frac{1}{6}\right) $$
    Since all the terms in the sum are identical, this simplifies to:
    $$ H(X) = -6 \cdot \left(\frac{1}{6} \log_2\left(\frac{1}{6}\right)\right) = -\log_2\left(\frac{1}{6}\right) \approx 2.585 $$

The entropy is approximately **2.585 bits**, which represents the average uncertainty of a single roll of a fair die.

# Optimisation

## Mini-Lesson: An Introduction to Optimization üéØ

**Optimization** is the process of finding the best set of parameters for a model in order to **minimize a cost function** (or maximize a reward function). In machine learning, this involves continuously adjusting model parameters to reduce the loss and improve accuracy.

---

### Gradient Descent
**Gradient descent** is a popular iterative optimization algorithm that updates model parameters in the direction that reduces error. It works by taking steps "downhill" on the loss function's surface.

The mathematical representation of the update rule is:
$$ \theta = \theta - \alpha \nabla J(\theta) $$
Where:
* $\theta$ is the parameter being optimized (e.g., a weight in a neural network).
* $J(\theta)$ is the loss function (or cost function).
* $\alpha$ is the **learning rate**, which determines the size of each step.
* $\nabla J(\theta)$ is the **gradient** (the derivative) of the loss function, which points in the direction of the steepest ascent. We move in the opposite direction to *descend*.

#### The Learning Rate ($\alpha$)
The learning rate is a critical hyperparameter that controls how large of a step we take during each iteration of gradient descent.
* If the learning rate is **too large**, the algorithm may overshoot the minimum and fail to converge.
* If it is **too small**, the algorithm will take a very long time to converge.

#### Worked Example
Let's use gradient descent to minimize the function $f(x) = (x-3)^2$. We'll start at the point $x_0 = 10$ with a learning rate of $\alpha = 0.1$.

* **Loss Function:** $J(x) = (x-3)^2$
* **Gradient:** $\nabla J(x) = 2(x-3)$
* **Update Rule:** $x_{\text{new}} = x_{\text{old}} - 0.1 \times (2(x_{\text{old}}-3))$

**Iteration 1:**
* Start with $x_0 = 10$.
* Compute the gradient: $\nabla J(10) = 2(10 - 3) = 14$.
* Update the parameter: $x_1 = 10 - (0.1 \times 14) = 10 - 1.4 = 8.6$.

**Iteration 2:**
* Start with $x_1 = 8.6$.
* Compute the gradient: $\nabla J(8.6) = 2(8.6 - 3) = 11.2$.
* Update the parameter: $x_2 = 8.6 - (0.1 \times 11.2) = 8.6 - 1.12 = 7.48$.

**Iteration 3:**
* Start with $x_2 = 7.48$.
* Compute the gradient: $\nabla J(7.48) = 2(7.48 - 3) = 8.96$.
* Update the parameter: $x_3 = 7.48 - (0.1 \times 8.96) = 7.48 - 0.896 = 6.584$.

With each step, the value of $x$ gets closer to 3, which is the minimum of the function $f(x)$.

---

### Local vs. Global Minimum
A key challenge in optimization is the difference between a local and global minimum.
* A **local minimum** is a point where the cost function is lower than at all nearby points, but not necessarily the lowest point overall.
* The **global minimum** is the absolute lowest point of the cost function across its entire domain.

Gradient descent can sometimes get "stuck" in a local minimum, preventing it from finding the true global minimum, especially in complex, non-convex loss landscapes.

## Mini-Lesson: Common Loss Functions üìâ

A **loss function**, also called a cost or objective function, quantifies the **error** between a model's predicted values and the true outcomes. The goal during model training is to minimize this function, which helps optimization algorithms like gradient descent adjust parameters and improve model accuracy.

---

### Mean Squared Error (MSE)
**Mean Squared Error** is the go-to loss function for **regression problems**, where the goal is to predict a continuous numerical value. It calculates the average of the squared differences between the predicted and actual values.

The formula is:
$$ MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$
Where:
* $y_i$ is the actual value.
* $\hat{y}_i$ is the predicted value.
* $n$ is the number of data points.

MSE is sensitive to outliers because the errors are squared, which gives more weight to larger errors. It provides a smooth, differentiable function, making it well-suited for gradient-based optimization.

#### MSE Example
Use the MSE formula for the following actual and predicted values:
* Actual $y = [3, 5, 6, 8]$
* Predicted $\hat{y} = [2.5, 4.8, 6.1, 7.9]$

**Calculation:**
$$ MSE = \frac{1}{4} [(3-2.5)^2 + (5-4.8)^2 + (6-6.1)^2 + (8-7.9)^2] $$
$$ MSE = \frac{1}{4} [(0.5)^2 + (0.2)^2 + (-0.1)^2 + (0.1)^2] $$
$$ MSE = \frac{1}{4} [0.25 + 0.04 + 0.01 + 0.01] = \frac{0.31}{4} = 0.0775 $$

---

### Cross-Entropy Loss (Log Loss)
**Cross-Entropy Loss** is used for **classification tasks**. It measures the difference between two probability distributions: the actual labels and the model's predicted probabilities. It penalizes incorrect predictions more heavily when the model was very confident.

#### Formulas
* **For binary classification:**
    $$ Loss = -\frac{1}{n} \sum_{i=1}^{n} [y_i \log(\hat{y}_i) + (1-y_i) \log(1-\hat{y}_i)] $$
* **For multi-class classification:**
    $$ Loss = -\sum_{i=1}^{n} \sum_{j=1}^{k} y_{ij} \log(\hat{y}_{ij}) $$
Where:
* $y_i$ is the actual label (0 or 1).
* $\hat{y}_i$ is the predicted probability.
* $n$ is the number of samples and $k$ is the number of classes.

#### Log Loss Example
Compute the log loss for three samples with the following labels and probabilities:

| Sample | Actual label ($y$) | Predicted probability ($\hat{y}$) |
| :--- | :--- | :--- |
| 1 | 1 | 0.9 |
| 2 | 0 | 0.2 |
| 3 | 1 | 0.4 |

**Calculation:**
* **Sample 1:** Loss = $-[1 \cdot \log(0.9) + 0 \cdot \log(0.1)] = -\log(0.9) \approx 0.1054$
* **Sample 2:** Loss = $-[0 \cdot \log(0.2) + 1 \cdot \log(0.8)] = -\log(0.8) \approx 0.2231$
* **Sample 3:** Loss = $-[1 \cdot \log(0.4) + 0 \cdot \log(0.6)] = -\log(0.4) \approx 0.9163$

**Average Log Loss:**
$$ \text{Loss} = \frac{0.1054 + 0.2231 + 0.9163}{3} \approx 0.4149 $$

## Mini-Lesson: Activation Functions üß†

An **activation function** is a mathematical function applied to a neuron (or node) in a neural network that determines its output. Its main purpose is to introduce **non-linearity** into the network, which is crucial for enabling the model to learn and solve complex problems. Without it, a neural network would behave like a simple linear model.

---

### Sigmoid Activation Function
The Sigmoid function maps any input value to a range between **0 and 1**. This makes it particularly useful in the output layer for **binary classification** tasks or for representing probabilities.

**Formula:**
$$ \sigma(x) = \frac{1}{1+e^{-x}} $$
**Properties:**
* The graph is an "S-shaped" curve.
* It's centered at $x=0$, where $\sigma(0) = 0.5$.
* As $x \to \infty$, $\sigma(x) \to 1$. As $x \to -\infty$, $\sigma(x) \to 0$.

**Drawback:** Sigmoid can suffer from the **vanishing gradients** problem, where gradients become extremely small during backpropagation, causing learning in deeper layers to become very slow or stop altogether.

---

### Rectified Linear Unit (ReLU)
The ReLU function is one of the most widely used activation functions in deep learning. It is much simpler computationally than Sigmoid.

**Formula:**
$$ f(x) = \max(0, x) $$
**Properties:**
* It outputs **0 for all negative inputs**.
* It outputs the input value directly (a **linear** function) for all positive inputs.
* It helps mitigate the vanishing gradient problem for positive values.

**Drawback:** ReLU can suffer from the **"dying ReLU" problem**, where a neuron gets stuck in the negative region and continuously outputs zero, effectively becoming inactive.

---

### Other Common Activation Functions
* **Tanh (Hyperbolic Tangent):** Similar to sigmoid but its output ranges from **-1 to 1**. This zero-centered output can sometimes help speed up convergence.
* **Leaky ReLU:** A variant of ReLU that allows a small, non-zero gradient for negative inputs. This is designed to prevent the "dying ReLU" problem.
* **Softmax:** Used exclusively in the output layer for **multi-class classification**. It converts a vector of raw scores (logits) into a probability distribution where the sum of all probabilities is 1.

### Worked Examples
Given the inputs $x = [-2, -1, 0, 1, 2]$, let's calculate the output for Sigmoid and ReLU.

#### 1. Sigmoid Function
* $\sigma(-2) = \frac{1}{1+e^{2}} \approx 0.1192$
* $\sigma(-1) = \frac{1}{1+e^{1}} \approx 0.2689$
* $\sigma(0) = \frac{1}{1+e^{0}} = 0.5$
* $\sigma(1) = \frac{1}{1+e^{-1}} \approx 0.7311$
* $\sigma(2) = \frac{1}{1+e^{-2}} \approx 0.8808$

**Sigmoid outputs:** `[0.1192, 0.2689, 0.5, 0.7311, 0.8808]`

#### 2. ReLU Function
* ReLU(-2) = $\max(0, -2) = 0$
* ReLU(-1) = $\max(0, -1) = 0$
* ReLU(0) = $\max(0, 0) = 0$
* ReLU(1) = $\max(0, 1) = 1$
* ReLU(2) = $\max(0, 2) = 2$

**ReLU outputs:** `[0, 0, 0, 1, 2]`
